```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("/data1/LLM_Models/Meta/llama2/Llama-2-7b", output_hidden_states=True,trust_remote_code=True)
```

```python
text = "救援队从废墟中救出五人，其中包括两名儿童。"
token_ids = tokenizer.encode(text)
tokens = tokenizer.tokenize(text)
print(tokens)
decoded_tokens = [tokenizer.decode([token_id]) for token_id in token_ids]
print(decoded_tokens)
```
